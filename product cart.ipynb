```python
import pymysql as p
```

```python
def add_cart(query):
    conn = p.connect(user="root", password="", host="localhost", database="product_cart")
    cur = conn.cursor()
    cur.execute(query)
    conn.commit()
    conn.close()

def show(query):
    conn = p.connect(user="root", password="", host="localhost", database="product_cart")
    cur = conn.cursor()
    cur.execute(query)
    conn.commit()
    records = cur.fetchall()
    conn.close()
    return records




while (True):
    print()
    print("Select any 1 of the options from below")
    choice = int(input(
        "1) ADD TO CART \n2) GENERATE BILL \n3) SHOW INVENTORY \n9) EXIT \nYOUR CHOICE : "))
    print()

# ADD TO CART
    if (choice == 1):
        P_ID = int(input("Enter Product ID : "))
        P_QTY = int(input("Enter Product Quantity : "))

        query = f"select * from inventory where P_ID={P_ID}"
        Get_Details = show(query)
        for i in Get_Details:
            P_NAME = i[1]
            P_COST = i[2]
            P_STOCK = i[3]
        UPDATED_INVENTORY_STOCK = P_STOCK - P_QTY

        query = f"select * from cart where P_ID={P_ID}"
        Get_Details = show(query)

        C_ID = ""
        C_STOCK = 0
        for i in Get_Details:
            C_ID = i[0]
            C_STOCK = i[3]
        UPDATED_CART_STOCK = P_QTY + C_STOCK


        if (P_STOCK >= P_QTY):

            # print(Total_Amout)

            if(C_ID == ""):
                Total_Amout = P_QTY * P_COST
                query = f"insert into cart (P_ID,P_NAME,P_COST,P_STOCK,Total_Amount) values ({P_ID},'{P_NAME}',{P_COST},{P_QTY},{Total_Amout})"
                add_cart(query)
                print("Item aadded in cart.......")
            else:
                Total_Amout = UPDATED_CART_STOCK * P_COST
                query = f"update cart set P_STOCK = {UPDATED_CART_STOCK},Total_Amount = {Total_Amout} where P_ID = {P_ID}"
                add_cart(query)
                print("Item aadded in cart.......")


            query = f"update inventory set P_STOCK={UPDATED_INVENTORY_STOCK} where P_ID={P_ID}"
            add_cart(query)
            print("Inventory Updated.......")


        else:
            print(f"Sock Left : {P_STOCK}")

# GENERATE BILL
    elif (choice == 2):

        query = f"select *  from cart"
        records = show(query)
        print("******************************")
        print("***  Showing Cart Details  ***")
        print("******************************")

        for j in records:
            print(f"P_ID : {j[0]}")
            print(f"P_NAME : {j[1]}")
            print(f"P_COST : {j[2]}")
            print(f"P_STOCK : {j[3]}")
            print(f"Total_Amount : {j[4]}")
            print("-----------x--------------")



# SHOW INVENTORY
    elif (choice == 3):
        query = f"select *  from inventory"
        print("***********************************")
        print("***  Showing Inventory Details  ***")
        print("***********************************")
        records = show(query)

        for j in records:
            print (f"P_ID : {j[0]}")
            print (f"P_NAME : {j[1]}")
            print (f"P_COST : {j[2]}")
            print (f"P_STOCK : {j[3]}")
            print("-----------x--------------")


# EXIT
    elif (choice == 9):
        break

    else:
        print("Invalid Input, Please Try Again .....")
```
