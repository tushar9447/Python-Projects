```python
#step1:  import pymysql
import pymysql as p

def insert_data(t):
    #step2: establish a connection with ur database 
    con=p.connect(user="root",password="",host="localhost",database="Employee")

    #step3: to create cursor object
    cur=con.cursor()  #it helps us to execute sql queries by providing different methods.
    
    query = "insert into Employee (Name,Department,Salary) values (%s,%s,%s)"
    
    cur.execute(query,t)
    
    con.commit()
    print("records inserted.......")
    
    con.close()
    
def update_data(t):
    con=p.connect(user="root",password="",host="localhost",database="Employee")

    cur=con.cursor()  
   
    query="update Employee set Name=%s , Department=%s , Salary=%s where ID=%s"
    
    cur.execute(query,t)
    
    con.commit() 
    
    print("record updated......")
    
    con.close()

def display_data():
    con=p.connect(user="root",password="",host="localhost",database="Employee")

   
    cur=con.cursor()
    
    query="select * from Employee"
    
    cur.execute(query)
    
    data=cur.fetchall() #FETCH THE RECORDS FROM CURSOR OBJECT
    for i in data:
        print("ID=",i[0])
        print("Name=",i[1])
        print("Department=",i[2])
        print("Salary=",i[3])
        print("---------x----------")

    con.commit() 
    
    con.close()

def delete_data(t):
    con=p.connect(user="root",password="",host="localhost",database="Employee")

    #step3: to create cursor object
    cur=con.cursor()
    
    query="delete from Employee where ID=%s"
    
    cur.execute(query,t)
    
    con.commit()
    
    print("-----Records Deleted Successfully---------------")
    con.close()
while True:    
        choice=int(input("press1: Insert Employee Data \npress2: Update Employee Data \npress3: Display Employee Data \npress4: Delete Employee Data \npress5 Exit \n:"))
        if choice==1:
            Name=input("Enter your Name")
            Department=input("Enter your Department")
            Salary=int(input("Enter your Salary"))
            t=(Name,Department,Salary)
            insert_data(t)

        if choice==2:
            print("------------WELCOME TO UPDATE PAGE---------------")
            ID=int(input("Enter your Employee ID"))
            Name=input("Enter your Name")
            Department=input("Enter your Department")
            Salary=int(input("Enter your Salary"))
            t=(Name,Department,Salary,ID)
            update_data(t)

        elif choice==3:
            display_data()

        elif choice==4:
            ID=int(input("Enter your Employee ID"))
            t=(ID,)
            delete_data(t)
        elif choice==5:
            break
        
```
